```python

# Importing necessary libraries
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#membaca dataset
housing = pd.read_csv('housing.csv')

#Pemisahan dataset menjadi data latih (80%) dan data uji (20%) menggunakan pengambilan sampel acak
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
test_set.head()

X_train = train_set.drop("median_house_value", axis=1)
y_train = train_set["median_house_value"]

# Data preprocessing pipeline
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])

num_features = list(X_train.select_dtypes(include=[np.number]))

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_features)
])

# Preprocess the training data
X_train_prepared = full_pipeline.fit_transform(X_train)
```

```python
# membangun model
model = Sequential([
    Dense(64, activation='relu', input_shape=[X_train_prepared.shape[1]]),
    Dense(32, activation='relu'),
    Dense(1)
])

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# melatih model
history = model.fit(X_train_prepared, y_train, epochs=50, batch_size=32, validation_split=0.2, verbose=1)
```
